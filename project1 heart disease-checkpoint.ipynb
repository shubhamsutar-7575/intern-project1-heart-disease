## Domain Analysis

1. heart_disease_present:
    Binary indicator (0/1) if heart disease is present.
2. patient_id:
    Unique identifier for each patient.
3. slope_of_peak_exercise_st_segment: 
    Slope during peak exercise (1-3); indicates heart stress.
    1 = Upsloping, 2 = Flat, 3 = Downsloping
4. thal:
    Type of blood defect i.e. results of thallium stress test(normal, fixed_defect,        reversible_defect).
5. resting_blood_pressure: 
    Patient’s blood pressure. A normal resting blood pressure for most adults is           between 90/60 and 120/80 millimeters of mercury (mmHg).
6. chest_pain_type:
    Type of chest pain which are encoded from (1-4). 
7. num_major_vessels: 
    Number of blocked vessels (0-3).
    0=Peripheral artery disease
    1=Carotid artery disease
    2=Coronary artery disease (CAD)
    3=Intestinal ischemic syndrome
8. fasting_blood_sugar_gt_120_mg_per_dl: 
    Fasting blood sugar >120 mg/dL. Here 0 means data is less than 120 and 1 means        greater than 120.
9. resting_ekg_results: 
    Resting electrocardiogram results (0-2).
    0=Normal EKG 
    1=Abnormal EKG
    2=Definite myocardial infarction (heart attack)
10. serum_cholesterol_mg_per_dl:
    Cholesterol level in the blood (mg/dL).
    It is an important indicator for assessing the risk of cardiovascular diseases.
11. oldpeak_eq_st_depression: 
    ST depression during exercise.
    0 mm: No ST depression, indicating normal results.
    1-2 mm: Mild to moderate ST depression, possibly indicating some level of             ischemia.
    >2 mm: Significant ST depression, indicating severe ischemia or higher risk of         heart disease.
12. sex: 
    Sex of the patient (0 = Female, 1 = Male).
13. age: 
    Age of the patient.
14. max_heart_rate_achieved: 
    Maximum heart rate.
15. exercise_induced_angina: 
    Whether chest pain occurred during exercise (0=flase,1=true).

## Model Selection

```python
x=data.drop('heart_disease_present',axis=1)
y=data.heart_disease_present
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=34,test_size=0.20)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# There is no need of smoting but for practice we use it
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_smote,y_smote=smote.fit_resample(x_train,y_train)
```

## Import The Model  --->

## Logistic Regression

```python
# We use logistic regression here
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,classification_report
```

```python
print('accuracy score:',accuracy_score(y_test,y_predict))
print('f1 score:',f1_score(y_test,y_predict))
print('classification report:',classification_report(y_test,y_predict))
```

## K-Nearest Neighbor

```python
# we randomly take one value of k
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
y_predict_knn=knn.predict(x_test)
print(accuracy_score(y_test,y_predict_knn))
print(f1_score(y_test,y_predict_knn))
print(classification_report(y_test,y_predict_knn))
```

```python
error_rate=[]
for i in range(1,11):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_smote,y_smote)
    pred_i=knn.predict(x_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(10,10),facecolor='white')
plt.plot(range(1,11),error_rate,linestyle='dashed',marker='o',markersize=10,color='red')
plt.title('error_rate vs. k.value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_smote,y_smote)
y_pred=knn.predict(x_test)
```

```python
print(f1_score(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
